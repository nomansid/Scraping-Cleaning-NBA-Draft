```python
import pandas as pd
import numpy as np

# 'cell magic' required to plot within IPython
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

draft_df = pd.read_csv("draft_data_1966_to_2015.csv", index_col=0)

draft_df.head()
```

```python
WS48_per_year = draft_df.groupby('Draft_Yr').WS_per_48.mean()
WS48_per_year
```

```python
# plot graph of draft data from 1966-2015
def plot_data(title, y_label, y_values, y_llim, y_ulim, source_space):
    sns.set_style("white")
    plt.figure(figsize=(12,9))
    x_values = draft_df.Draft_Yr.unique()
    plt.title(title, fontsize=20)
    plt.ylabel(y_label, fontsize=18)
    plt.xlim(1966,2015.5)
    plt.ylim(y_llim,y_ulim)
    plt.grid(axis='y', color='grey', linestyle='--', lw=0.5, alpha=0.5)
    plt.tick_params(axis='both', labelsize=14)
    sns.despine(left=True, bottom=True)
    plt.plot(x_values, y_values)
    plt.text(1966, source_space, 
         'Primary Data Source: http://www.basketbal-reference.com/draft/'
         '\nAuthor: Noman Siddiqui (nomansid.github.io)',
         fontsize=12)
    plt.show()

# Plotting WS/48 by year
plot_data('Average Career Win Shares Per 48 minutes by Draft Year (1966-2015)',
         'Win Shares Per 48 minutes', 
          draft_df.groupby('Draft_Yr').WS_per_48.mean(),
         0,0.08,-0.012)
```

```python
# Plot number of players drafter in each draft
plot_data('The Number of Players Drafted in each Draft (1966-2015)',
         'Number of Players Drafted', draft_df.groupby('Draft_Yr').Pk.count(),
         0, 250, -35)
```
